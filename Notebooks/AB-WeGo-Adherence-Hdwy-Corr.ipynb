```python
ishdwy_dev_normal = st.normaltest(headway_df['HDWY_DEV'][headway_df['HDWY_DEV'].notnull()])
ishdwy_dev_normal_pvalue=ishdwy_dev_normal[1]

if ishdwy_dev_normal_pvalue>=0.05:
    print('The p value was {:.3f}. The null hypothesis of the normality test was not rejected. The headway deviation distribution can be assumed to be normal.'.format(ishdwy_dev_normal_pvalue))
else:
    print('The p value was {:.3f}. The null hypothesis of the normality test was rejected. The headway deviation distribution cannot be assumed to be normal.'.format(ishdwy_dev_normal_pvalue))
```

```python
isadherence_normal = st.normaltest(headway_df['ADHERENCE'][headway_df['ADHERENCE'].notnull()])
isadherence_normal_pvalue=isadherence_normal[1]

if isadherence_normal_pvalue>=0.05:
    print('The p value was {:.3f}. The null hypothesis of the normality test was not rejected. The headway deviation distribution can be assumed to be normal.'.format(isadherence_normal_pvalue))
else:
    print('The p value was {:.3f}. The null hypothesis of the normality test was rejected. The headway deviation distribution cannot be assumed to be normal.'.format(isadherence_normal_pvalue))
```

```python
headway_df['LOG_ADHERENCE']= np.log(headway_df['ADHERENCE'])
headway_df['LOG_HDWY_DEV'] = np.log(headway_df['HDWY_DEV'])
headway_df['LOG_SCHEDULED_HDWY'] = np.log(headway_df['SCHEDULED_HDWY'])
headway_df['LOG_ACTUAL_HDWY'] = np.log(headway_df['ACTUAL_HDWY'])
```
